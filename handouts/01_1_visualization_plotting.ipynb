## With Noise

```python
N = 1000
signal_a = np.linspace(-100, 100, N)
```

```python
plt.hist(signal_a, bins=15)
```

* [Uniform distribution (continuous) – Wikipedia](https://en.wikipedia.org/wiki/Uniform_distribution_%28continuous%29)

```python
# loc: μ, mu, mean
# scale: σ, sigma, standard deviation
# size: the length of the ndarray
# sample from the normal distribution
noise_a = np.random.normal(loc=0, scale=10, size=N)
```

```python
plt.hist(noise_a, bins=15)
```

* [Normal distribution – Wikipedia](https://en.wikipedia.org/wiki/Normal_distribution)


### [Central limit theorem – Wikipedia](https://en.wikipedia.org/wiki/Central_limit_theorem)

When independent random variables are added, their sum tends toward a normal distribution even if the original variables are not normally distributed.
    
$ {\displaystyle {\sqrt {n}}\left(\left({\frac {1}{n}}\sum _{i=1}^{n}X_{i}\right)-\mu \right)\ {\xrightarrow {d}}\ N\left(0,\sigma ^{2}\right).} $

* $ n $: approaches infinity.
* $ X_i $: a random variable.
* $ {\xrightarrow {d}} $: [converge in distribution](https://en.wikipedia.org/wiki/Convergence_in_distribution).
* $ N $: normal distribution.
* $ \mu $: mean of the random variables.
* $ \sigma^2 $: variance of the random variables.

```python
sn_a = signal_a + noise_a
```

```python
plt.scatter(x=signal_a, y=sn_a, alpha=0.1)
```

```python
plt.plot(signal_a, signal_a)
plt.scatter(x=signal_a, y=sn_a, alpha=0.1)
```

## Put Some Texts

```python
N = 1000
a = np.linspace(-100, 100, N)

plt.title('My Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(a, a**2)
plt.text(
    x=90, y=300,
    s=r'$y = x^2$', ha='center', size=15
)
plt.annotate(
    xy=(0, 0),
    arrowprops=dict(facecolor='red', shrink=0.05),
    xytext=(0, 2000),
    s='min', ha='center', size=15,
)
plt.legend(['My Line'])
```
