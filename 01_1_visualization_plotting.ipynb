```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(font='Noto Sans CJK TC')
mpl.style.use('ggplot')  # must put after sns.set
```

```python
plt.scatter(x=[1, 2, 3, 4, 5, 6], y=[1, 4, 9, 16, 25, 36])
```

```python
N = 50
sample_a = np.linspace(-100, 100, N)
```

```python
plt.scatter(x=sample_a, y=sample_a)
```

```python
sample_with_noise_a = sample_a + np.random.normal(loc=0, scale=10, size=N)
```

```python
plt.scatter(x=sample_a, y=sample_with_noise_a)
```

```python
plt.plot(sample_a, sample_a**2)
```

```python
plt.plot(sample_a, 2**sample_a)
```

```python
plt.plot(sample_a, np.log2(sample_a))
plt.plot(sample_a, np.log(sample_a))
plt.plot(sample_a, np.log10(sample_a))
```

```python
plt.title('My Plot')
plt.plot(sample_a, sample_a**2)
plt.text(
    x=90, y=300,
    s=r'$y = x^2$', ha='center', size=15
)
plt.annotate(
    xy=(0, 0),
    arrowprops=dict(facecolor='red', shrink=0.05),
    xytext=(0, 2000),
    s='min', ha='center', size=15,
)
```
